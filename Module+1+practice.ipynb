```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

fruits=pd.read_csv("fruit_data_with_colors.csv")
```

```python
fruits.head()
```

```python
fruits
```

```python
fruits.shape
```

```python
lookup_fruit_name=dict(zip(fruits.fruit_label.unique(),fruits.fruit_name.unique()))
```

```python
lookup_fruit_name
```

```python
x=fruits[['mass','width','height','color_score']]
y=fruits['fruit_label']
#x
x_train, x_test, y_train, y_test=train_test_split(x,y,random_state=0)
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from matplotlib import cm
cmap=cm.get_cmap("gnuplot")
scatter=pd.scatter_matrix(x_train,alpha=0.5,c=y_train,diagonal='kde',marker='o',s=40,hist_kwds={'bins':15},figsize=(12,12),cmap=cmap,grid=True)
plt.show()
```

```python
from matplotlib import cm
cmap=cm.get_cmap("gnuplot")
scatter=pd.scatter_matrix(x_train,c=y_train,alpha=1,marker='o',hist_kwds={'bins':15},s=100,figsize=(12,12),diagonal='hist',cmap=cmap)
plt.show()
```

```python
#create classifier object
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
 knn.score(x_test,y_test)
```

```python
fruit_prediction = knn.predict([[190,8.7,7.3,0.55]])
lookup_fruit_name[fruit_prediction[0]]
```
